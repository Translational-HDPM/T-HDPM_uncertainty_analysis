```python
import pandas as pd
MCfolder = '../MC-results-S2/'
files = pd.Series(os.listdir(MCfolder))
csvs = files.str.split('-',expand = True).iloc[:,4:10]
csvs.columns = ['Filter','Runs','Sigma','Run','Blank','Dist']
```

```python
csvs = pd.concat((csvs,pd.concat([pd.read_csv(os.path.join(MCfolder,csv)) for csv in files]).reset_index(drop=True)),axis = 1)
csvs
```

```python
fig,ax = plt.subplots(figsize = (10,10))
csvs = csvs.sort_values(by=['Filter'])
plt.errorbar(csvs[csvs.Dist == 'LHS'].Filter,csvs[csvs.Dist == 'LHS']['AD Group Mean'],csvs[csvs.Dist == 'LHS']['AD Group Standard Deviation'])
plt.errorbar(csvs[csvs.Dist == 'gamma'].Filter,csvs[csvs.Dist == 'gamma']['AD Group Mean'],csvs[csvs.Dist == 'gamma']['AD Group Standard Deviation'])
plt.errorbar(csvs[csvs.Dist == 'rayleigh'].Filter,csvs[csvs.Dist == 'rayleigh']['AD Group Mean'],csvs[csvs.Dist == 'rayleigh']['AD Group Standard Deviation'])

# set labels
ax.set_xlabel('Filter', fontsize=25, fontweight='black', color = '#333F4B',fontname = 'Arial')
ax.set_ylabel('AD Group Mean', fontsize=25, fontweight='black', color = '#333F4B',fontname = 'Arial')

# set axis
tickfont = {'fontname':'Arial','fontsize':16}
plt.yticks(**tickfont)
plt.xticks(**tickfont)
ax.tick_params(axis="y",direction="in")

# set grids
grid_setting = {'linestyle':'--', 'linewidth':2,'alpha':0.4}
plt.grid(which='major', axis='y', **grid_setting)
plt.legend(['LHS','gamma','rayleigh'])
fig,ax = plt.subplots(figsize = (10,10))

plt.errorbar(csvs[csvs.Dist == 'LHS'].Filter,csvs[csvs.Dist == 'LHS']['NCI Group Mean'],csvs[csvs.Dist == 'LHS']['NCI Group Standard Deviation'])
plt.errorbar(csvs[csvs.Dist == 'gamma'].Filter,csvs[csvs.Dist == 'gamma']['NCI Group Mean'],csvs[csvs.Dist == 'gamma']['NCI Group Standard Deviation'])
plt.errorbar(csvs[csvs.Dist == 'rayleigh'].Filter,csvs[csvs.Dist == 'rayleigh']['NCI Group Mean'],csvs[csvs.Dist == 'rayleigh']['NCI Group Standard Deviation'])

# set labels
ax.set_xlabel('Filter', fontsize=25, fontweight='black', color = '#333F4B',fontname = 'Arial')
ax.set_ylabel('NCI Group Mean', fontsize=25, fontweight='black', color = '#333F4B',fontname = 'Arial')

# set axis
tickfont = {'fontname':'Arial','fontsize':16}
plt.yticks(**tickfont)
plt.xticks(**tickfont)
ax.tick_params(axis="y",direction="in")

# set grids
grid_setting = {'linestyle':'--', 'linewidth':2,'alpha':0.4}
plt.grid(which='major', axis='y', **grid_setting)
plt.legend(['LHS','gamma','rayleigh'])
```
