```python
# sort the columns of Patients_df in ascending order based on their index values
Patients_df.sort_index(axis=1, inplace=True)
Patient_r1.sort_index(axis=1, inplace=True)
Patient_r2.sort_index(axis=1, inplace=True)
```

```python
# Create a list patient_id with column names in Patients_df
patient_id= list(Patients_df.columns.values)
patient_r1= list(Patient_r1.columns.values)
patient_r2= list(Patient_r2.columns.values)
#patient_id
#patient_r1
#patient_r2
```

```python
Patient_df = Patients_df.copy()
```

```python
# Calculate Mean, Std and RSD for Patients_df and add respective columns
Patient_df['Mean']= Patient_df.mean(axis=1)
Patient_df['Std']=Patient_df.iloc[:,:-1].std(axis=1)
Patient_df['RSD'] = (Patient_df['Std'] / Patient_df['Mean'])*100 # New code Filip

Patient_df.head()
#print(Patient_df.shape())
```

```python
# Calculate Mean, Std and RSD for Patients_r1 and add respective columns
Patient_r1['Mean_r1']= Patient_r1.mean(axis=1)
Patient_r1['Std_r1']= Patient_r1.iloc[:,:-1].std(axis=1)
Patient_r1['RSD_r1'] = (Patient_r1['Std_r1'] /Patient_r1['Mean_r1'])*100 # New code Filip

Patient_r1.head()
#print(Patients_r1.shape)
```

```python
# Calculate Mean, Std and RSD for Patients_r2 and add respective columns
Patient_r2['Mean_r2']= Patient_r2.mean(axis=1)
Patient_r2['Std_r2']= Patient_r2.iloc[:,:-1].std(axis=1)
Patient_r2['RSD_r2'] = (Patient_r2['Std_r2'] /Patient_r2['Mean_r2'])*100 # New code Filip

Patient_r2.head()
#print(Patients_r1.shape)
```

```python
# We define a function whose input is TPM and outputs the corresponding Zscore
def z_score(x):
    return (x-x['Mean'])/x['Std']
```

```python
# Computing and storing zscores for Patient_df
Patient_df_zScore = Patient_df.apply(lambda x: z_score(x), axis=1)
Patient_df_zScore.head()
#print(Patients_df_zScore.shape)
```

```python
# We define a function whose input is TPM and outputs the corresponding Zscore for Patient_r1
def z_score_r1(x):
    return (x-x['Mean_r1'])/x['Std_r1']
```

```python
# Computing and storing zscores for Patient_r1
Patient_r1_zScore = Patient_r1.apply(lambda x: z_score_r1(x), axis=1)
Patient_r1_zScore.head()
#print(Patients_df_zScore.shape)
```

```python
# We define a function whose input is TPM and outputs the corresponding Zscore for Patient_r2
def z_score_r2(x):
    return (x-x['Mean_r2'])/x['Std_r2']
```

```python
# Computing and storing zscores for Patient_r2
Patient_r2_zScore = Patient_r2.apply(lambda x: z_score_r2(x), axis=1)
Patient_r2_zScore.head()
#print(Patients_df_zScore.shape)
```

```python
Patient_df_zScore.reset_index(inplace=True)
Patient_r1_zScore.reset_index(inplace=True)
Patient_r2_zScore.reset_index(inplace=True)
```

```python
Patient_df_zScore.head()
```

```python
gene_scores = Patient_df_zScore.loc[:, ['gene_id', 'Std', 'RSD']]
gene_scores_r1 = Patient_r1_zScore.loc[:, ['gene_id', 'Std_r1', 'RSD_r1']]
gene_scores_r2 = Patient_r2_zScore.loc[:, ['gene_id','Std_r2', 'RSD_r2']]
```

```python
gene_scores_r2.head()
```

```python
gene_scores_all = gene_scores.merge(gene_scores_r1, on='gene_id').merge(gene_scores_r2, on='gene_id')
```
