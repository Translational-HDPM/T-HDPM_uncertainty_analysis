```python
import pandas as pd
import matplotlib.pyplot as plt
import os 
# Font
plt.rcParams['font.family']     = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Helvetica'

# set the style of the axes and the text color
plt.rcParams["axes.labelweight"] = "bold"
plt.rcParams["font.weight"] = "normal"
plt.rcParams['axes.edgecolor']='#333F4B'
plt.rcParams['axes.linewidth']=1.5
plt.rcParams['xtick.color']='#333F4B'
plt.rcParams['ytick.color']='#333F4B'
plt.rcParams['text.color']='#333F4B'
csfont = {'fontname':'Helvetica','fontsize':18}
```

```python
files = pd.Series(os.listdir('MC-results-S4/'))
results = files[files.str.contains('results')]
simruns = files[files.str.contains('simruns')]
tpmresults = results[results.str.contains('Avg')]
coeffresults = results[results.str.contains('Coeff')]
smallestcoeffresults = coeffresults[coeffresults.str.contains('smallest')]
largestcoeffresults = coeffresults[coeffresults.str.contains('largest')]
smallesttpmresults = tpmresults[tpmresults.str.contains('smallest')]
largesttpmresults = tpmresults[tpmresults.str.contains('largest')]


bothcoeffresults = coeffresults[coeffresults.str.contains('both')]
bothtpmresults = tpmresults[tpmresults.str.contains('both')]
```

## nsmallest

```python
csvs = smallesttpmresults.sort_values().tolist()
df = pd.concat([pd.read_csv('MC-results-S4/'+csv) for csv in csvs],ignore_index = True)
df['filter'] = [0,100,200,400,800]
df

csvs = smallestcoeffresults.sort_values().tolist()
df1 = pd.concat([pd.read_csv('MC-results-S4/'+csv) for csv in csvs],ignore_index = True)
df1['filter'] = [0,100,200,400,800]
```

```python
fig,ax = plt.subplots(figsize = (10,10))
plt.errorbar(df['filter'],df['AD Group Mean'],df['AD Group Standard Deviation'],label = 'Coeff * TPM')
plt.errorbar(df1['filter'],df1['AD Group Mean'],df1['AD Group Standard Deviation'],label = 'TPM')

# set labels
ax.set_xlabel('Filter', fontsize=25, fontweight='black', color = '#333F4B',fontname = 'Arial')
ax.set_ylabel('TPM', fontsize=25, fontweight='black', color = '#333F4B',fontname = 'Arial')
plt.legend(title = 'Column',fontsize = 15)
# set axis
tickfont = {'fontname':'Arial','fontsize':16}
plt.yticks(**tickfont)
plt.xticks(**tickfont)
ax.tick_params(axis="y",direction="in")

# set grids
grid_setting = {'linestyle':'--', 'linewidth':2,'alpha':0.4}
plt.grid(which='major', axis='y', **grid_setting)


fig,ax = plt.subplots(figsize = (10,10))
plt.errorbar(df['filter'],df['NCI Group Mean'],df['NCI Group Standard Deviation'],label = 'Coeff * TPM')
plt.errorbar(df1['filter'],df1['NCI Group Mean'],df1['NCI Group Standard Deviation'],label = 'TPM')
plt.legend(title = 'Column',fontsize = 15)
# set labels
ax.set_xlabel('Filter', fontsize=25, fontweight='black', color = '#333F4B',fontname = 'Arial')
ax.set_ylabel('TPM', fontsize=25, fontweight='black', color = '#333F4B',fontname = 'Arial')

# set axis
tickfont = {'fontname':'Arial','fontsize':16}
plt.yticks(**tickfont)
plt.xticks(**tickfont)
ax.tick_params(axis="y",direction="in")

# set grids
grid_setting = {'linestyle':'--', 'linewidth':2,'alpha':0.4}
plt.grid(which='major', axis='y', **grid_setting)


fig,ax = plt.subplots(figsize = (10,10))
plt.plot(df['filter'],df['AD Group Mean']/df['NCI Group Mean'],label = 'Coeff * TPM')
plt.plot(df1['filter'],df1['AD Group Mean']/df['NCI Group Mean'],label = 'TPM')

# set labels
ax.set_xlabel('Filter', fontsize=25, fontweight='black', color = '#333F4B',fontname = 'Arial')
ax.set_ylabel('AD TPM/NCI TPM', fontsize=25, fontweight='black', color = '#333F4B',fontname = 'Arial')
plt.legend(title = 'Column',fontsize = 15)
# set axis
tickfont = {'fontname':'Arial','fontsize':16}
plt.yticks(**tickfont)
plt.xticks(**tickfont)
ax.tick_params(axis="y",direction="in")

# set grids
grid_setting = {'linestyle':'--', 'linewidth':2,'alpha':0.4}
plt.grid(which='major', axis='y', **grid_setting)
```
