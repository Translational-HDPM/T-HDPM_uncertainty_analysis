```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
```

```python
excelsheet = pd.ExcelFile('/anvil/projects/x-cis220051/corporate/molecular-stethoscope/Teams/Team-2/ClusterMarkers_1819ADcohort-Copy1.congregated_DR.xlsx')
TPMs = excelsheet.parse(excelsheet.sheet_names[-1])
TPMs = TPMs[~TPMs.Coeff.isna()]
df = pd.read_csv('AD-data.csv',index_col =0)
df = df.iloc[TPMs.gene_id.index.tolist(),:-6].T
df.head(5)
```

```python
Coeff = TPMs.Coeff
mins = TPMs.MIN
maxes = TPMs.MAX
genes = TPMs.gene_id
```

```python
excelsheet = pd.ExcelFile('/anvil/projects/x-cis220051/corporate/molecular-stethoscope/Teams/Team-2/ClusterMarkers_1819ADcohort-Copy1.congregated_DR.xlsx')
TPMs = excelsheet.parse(excelsheet.sheet_names[-1])
TPMs = TPMs[~TPMs.Coeff.isna()]
df1 = pd.read_csv('NCI-data.csv',index_col =0)
df1 = df1.iloc[TPMs.gene_id.index.tolist(),:-6].T
df1.head(5)
```

```python
excelsheet = pd.ExcelFile('/anvil/projects/x-cis220051/corporate/molecular-stethoscope/Teams/Team-2/ClusterMarkers_1819ADcohort-Copy1.congregated_DR.xlsx')
TPMs = excelsheet.parse(excelsheet.sheet_names[-1])
TPMs = TPMs[~TPMs.Coeff.isna()]
dfall = pd.read_csv('all-data.csv',index_col =0)
dfall = dfall.iloc[TPMs.gene_id.index.tolist(),:-6].T
dfall.head(5)

```

```python
metrics = dict()
metrics['gene'] = genes
metrics['AVG TPM AD'] = df.mean()
metrics['AVG TPM NCI'] = df1.mean()
metrics['AVG TPM all'] = dfall.mean()

metrics['MED TPM AD'] = df.median()
metrics['MED TPM NCI'] = df1.median()
metrics['MED TPM all'] = dfall.median()

metrics['AVG BTPM AD'] = np.multiply(df.mean(),Coeff)
metrics['AVG BTPM NCI'] = np.multiply(df1.mean(),Coeff)
metrics['AVG BTPM all'] = np.multiply(dfall.mean(),Coeff)

metrix ['MED BTPM AD'] = np.multiply(df.median(), Coeff)
metrics['MED BTPM NCI'] = np.multiply(df1.median(),Coeff)
metrics['MED BTPM all'] = np.multiply(dfall.median(),Coeff)

metrics['STD TPM AD'] = df.std()
metrics['STD TPM NCI'] = df1.std()
metrics['STD TPM all'] = dfall.std()

metrics['STD BTPM AD'] = np.multiply(df.std(),Coeff)
metrics['STD BTPM NCI'] = np.multiply(df1.std(),Coeff)
metrics['STD BTPM all'] = np.multiply(dfall.std(),Coeff)

metrics['B'] = Coeff



metrics['MIN TPM AD'] = df.min()
metrics['MIN TPM NCI'] = df1.min()
metrics['MIN TPM all'] = dfall.min()

metrics['MAX TPM AD'] = df.max()
metrics['MAX TPM NCI'] = df1.max()
metrics['MAX TPM all'] = dfall.max()

metrics = pd.DataFrame(metrics)
metrics.to_csv('Updated Toden Values.csv')
metrics.head(5)
```

```python
fig,ax = plt.subplots(figsize= (10,10))
plt.plot([0,7000],[0,7000],'--r',linewidth = 3)
plt.scatter(metrics['AVG TPM AD'], metrics['AVG TPM NCI'],color="black")
plt.errorbar(metrics['AVG TPM AD'], metrics['AVG TPM NCI'], xerr=metrics['STD TPM AD'],yerr=metrics['STD TPM NCI'], fmt="o",alpha = .5)

plt.grid()
plt.xlim([0,7000])
plt.ylim([0,7000])
plt.xlabel('AVG TPM AD')
plt.ylabel('AVG TPM NCI')

```
