```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family']     = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Helvetica'

# set the style of the axes and the text color
plt.rcParams["axes.labelweight"] = "bold"
plt.rcParams["font.weight"] = "normal"
plt.rcParams['axes.edgecolor']='#333F4B'
plt.rcParams['axes.linewidth']=1.5
plt.rcParams['xtick.color']='#333F4B'
plt.rcParams['ytick.color']='#333F4B'
plt.rcParams['text.color']='#333F4B'
csfont = {'fontname':'Helvetica','fontsize':18}
```

```python
df = pd.read_csv('all-data.csv')
bardata = df.groupby('Popular Dist').count()['Popular Score'].reset_index()
bardata.columns = ['Dist','all']
bardata = bardata.append(pd.DataFrame([['loggamma',0]],index=['6'],columns=bardata.columns))
bardata = bardata.sort_values('Dist').reset_index(drop=True)
df = pd.read_csv('AD-data.csv')
bardata['AD'] = df.groupby('Popular Dist').count()['Popular Score'].values
df = pd.read_csv('NCI-data.csv')
bardata['NCI'] = df.groupby('Popular Dist').count()['Popular Score'].values
bardata.plot.bar(x='Dist')
plt.xlabel('Distribution',**csfont)
tickfont = {'fontname':'Arial','fontsize':16}
plt.yticks(**tickfont)
plt.xticks(**tickfont)
plt.ylabel('Genes',**csfont)
plt.grid()
```

```python
df = pd.read_csv('AD-data-r2.csv')
bardata = df.groupby('Popular Dist').count()['Popular Score'].reset_index()
bardata.columns = ['Dist','AD-r2']
df = pd.read_csv('NCI-data-r2.csv')
bardata['NCI-r2'] = df.groupby('Popular Dist').count()['Popular Score'].values
df = pd.read_csv('all-data-r2.csv')
bardata['all-r2'] = df.groupby('Popular Dist').count()['Popular Score'].values

bardata.plot.bar(x='Dist')
plt.grid()
```
